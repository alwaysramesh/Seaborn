```python
Titanic=sns.load_dataset('titanic')
print(Titanic)
```

```python
Titanic.head(20)
```

```python
Titanic['age'].fillna(Titanic['age'].mean(),inplace=True)
```

```python
sns.distplot(Titanic['age'])
```

```python
sns.distplot(Titanic['age'],kde=True)
```

```python
sns.distplot(Titanic['age'],bins=10,kde=True)
```

```python
sns.distplot(Titanic['age'],hist = False)
plt.show()
```
