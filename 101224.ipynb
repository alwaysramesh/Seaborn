```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('Downloads/tips.csv')
print(df)
```

```python
sns.catplot(x='smoker',y='tip',kind='bar',hue='sex',data=df)
```

```python
sns.catplot(x='smoker',y='tip',kind='bar',hue='sex',estimator='mean',data=df)
plt.show()
```

```python
sns.catplot(x='smoker',y='tip',kind='bar',hue='sex',estimator='median',data=df)
```

```python
sns.catplot(x='smoker',y='tip',kind='bar',hue='sex',estimator='median',data=df)
```

```python
#countplot
sns.catplot(x='smoker',kind='count',data=df)
```

```python
flight=sns.load_dataset('flights')
print(flight)
```

```python
x=flight.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
print(x)
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x,cbar=False,linewidth=0.8)
```

```python
sns.heatmap(x,cbar=False,linewidth=0.6,annot=True,fmt='d')
```

```python
sns.heatmap(x,cbar=False,linewidth=0.6,annot=True,fmt='d',cmap='spring')
```

```python
sns.heatmap(x,cbar=False,linewidth=0.6,annot=True,fmt='d',cmap='autumn')
```

```python
sns.heatmap(x,cbar=False,linewidth=2,annot=True,fmt='d',linecolor='black',cmap='winter')
```

```python
sns.heatmap(x,cbar=False,linewidth=2,annot=True,fmt='d',linecolor='black',cmap='summer')
```

```python
#cluster map
x=flight.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
print(x)
```

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, z_score= 0 ,metric='correlation')
plt.show()
```

```python
sns.clustermap(x, z_score= 0 ,annot=True)
plt.show()
```
