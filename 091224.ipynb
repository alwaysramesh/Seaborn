```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('Downloads/tips.csv')
print(df)
```

```python
sns.boxplot(df['tip'])
plt.show()
#5 summary box bhi bhi kahte hai
```

```python
print(df.describe())
```

```python
sns.boxplot(df['total_bill'])
plt.show()
```

```python
#byvariant

sns.catplot(x='day',y='tip',kind='box',data=df)
plt.show()
```

```python
#byvariant

sns.catplot(x='sex',y='tip',kind='box',data=df)
plt.show()
```

```python
#byvariant

sns.catplot(x='size',y='tip',kind='box',data=df)
plt.show()
```

```python
#byvariant

sns.catplot(x='day',y='total_bill',kind='box',data=df)
plt.show()
```

```python
#byvariant

sns.catplot(x='day',y='tip',kind='box',hue='sex',data=df)
plt.show()
#decender order day
```

```python

sns.catplot(x='day',y='tip',kind='box',hue='smoker',data=df)
plt.show()
#decender order day
```
