```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('Downloads/tips.csv')
print(df)
```

```python
#tips dataset used

plt.style.use('fivethirtyeight')

sns.catplot(x='day',y='tip',kind='strip',data=df)
plt.show()
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=0,data=df)
plt.show()
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=df)
#plt.show()
```

```python
sns.catplot(x='day',y='total_bill',kind='swarm',data=df)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',kind='strip',jitter=100,data=df)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',kind='strip',hue='smoker',data=df)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',kind='strip',hue='size',data=df)
plt.show()
```
