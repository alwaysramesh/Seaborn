```python
df=pd.read_csv('Downloads/tips.csv')
print(df)
```

```python
print(df.describe())
```

```python
sns.distplot(df['total_bill'])
sns.distplot(df['total_bill'],kde=False)

```

```python
sns.distplot(df['total_bill'],kde=False,bins=30)
```

```python
sns.distplot(df['tip'],kde=False,bins=30)
```

```python
sns.distplot(df['tip'])
sns.distplot(df['tip'],kde=False)

```

```python
print(df['tip'].max())
```

```python
import numpy as np
plt.style.use('fivethirtyeight')
df.head()
```

```python

#scatter plot
sns.relplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python

#hue parameter
sns.relplot(x='total_bill',y='tip',hue='smoker',data=df)
plt.show()
```

```python
#style parameter
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=df)
plt.show()
```

```python
#size parameter
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=df)
plt.show()
```
